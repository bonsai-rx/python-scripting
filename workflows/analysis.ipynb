```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1_name = "baseline"
df1 = pd.read_csv(f"{df1_name}.csv", header = None)
df1["Time"] = pd.to_timedelta(df1[1])
df1
```

```python
df2_name = "baseline_wRuntime"
df2 = pd.read_csv(f"{df2_name}.csv", header = None)
df2["Time"] = pd.to_timedelta(df2[1])
df2
```

```python
df3_name = "baseline_wSample"
df3 = pd.read_csv(f"{df3_name}.csv", header = None)
df3["Time"] = pd.to_timedelta(df3[1])
df3
```

```python
df4_name = "baseline_test"
df4 = pd.read_csv(f"{df4_name}.csv", header = None)
df4["Time"] = pd.to_timedelta(df4[1])
df4
```

```python
df5_name = "test"
df5 = pd.read_csv(f"{df5_name}.csv", header = None)
df5["Time"] = pd.to_timedelta(df5[1])
df5
```

```python
start = 0
end = 10000
plt.plot(df1.index[start:end], df1.Time.dt.total_seconds()[start:end], label=df1_name, alpha = 0.3)
plt.plot(df2.index[start:end], df2.Time.dt.total_seconds()[start:end], label=df2_name, alpha = 0.3)
plt.plot(df3.index[start:end], df3.Time.dt.total_seconds()[start:end], label=df3_name, alpha = 0.3)
plt.plot(df4.index[start:end], df4.Time.dt.total_seconds()[start:end], label=df4_name, alpha = 0.3)
plt.plot(df5.index[start:end], df5.Time.dt.total_seconds()[start:end], label=df5_name, alpha = 0.3)
plt.ylim(0, 0.0005)
plt.legend()
```

```python
tmp_df = pd.concat([df1.Time, df2.Time, df3.Time, df4.Time, df5.Time], axis = 1, keys=[df1_name, df2_name, df3_name, df4_name, df5_name])
```
