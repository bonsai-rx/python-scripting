```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "baseline_wRuntime.csv"
df1 = pd.read_csv(filename, header = None)
df1["Time"] = pd.to_timedelta(df1[1])
df1
```

```python
filename = "baseline_test.csv"
df2 = pd.read_csv(filename, header = None)
df2["Time"] = pd.to_timedelta(df2[1])
df2
```

```python
filename = "baseline.csv"
df3 = pd.read_csv(filename, header = None)
df3["Time"] = pd.to_timedelta(df3[1])
df3
```

```python
start = 0
end = 10000
plt.plot(df1.index[start:end], df1.Time.dt.total_seconds()[start:end], label="baseline_wRuntime", alpha = 0.3)
plt.plot(df2.index[start:end], df2.Time.dt.total_seconds()[start:end], label="baseline_test", alpha = 0.3)
plt.plot(df3.index[start:end], df3.Time.dt.total_seconds()[start:end], label="baselinet", alpha = 0.3)
plt.ylim(0, 0.02)
```
